## Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('penguins')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='bill_length_mm', y='bill_depth_mm')
plt.title("Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.head(20), x='bill_length_mm', y='bill_depth_mm', hue='sex')
plt.title("Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.head(20), x='bill_length_mm', y='bill_depth_mm', hue='sex', style='sex')
plt.title("Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.head(20), x='bill_length_mm', y='bill_depth_mm', hue='sex', style='sex', size='sex')
plt.title("Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.head(20), x='bill_length_mm', y='bill_depth_mm', hue='sex', style='sex', size='sex',
               sizes=(100, 40))
plt.title("Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.head(20), x='bill_length_mm', y='bill_depth_mm', hue='sex', style='sex', size='sex',
               sizes=(100, 60), palette='gist_ncar')
plt.title("Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.head(20), x='bill_length_mm', y='bill_depth_mm', hue='sex', style='sex', size='sex',
               sizes=(100, 60), palette='gist_ncar', alpha=0.7)
plt.title("Scatter Plot")
plt.show()
```

```python
m = {"Male":"*", "Female":"o"}

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='bill_length_mm', y='bill_depth_mm', hue='sex', style='sex', size='sex',
               sizes=(100, 60), palette='gist_ncar', alpha=0.7, markers=m)
plt.title("Scatter Plot")
plt.show()
```
