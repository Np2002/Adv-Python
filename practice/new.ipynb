```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('penguins')
data.head()
```

```python
species_bill_length = data.groupby('island').agg({'bill_length_mm':'mean', 'bill_depth_mm':'mean', 'flipper_length_mm':'mean', 'body_mass_g':'mean'}).reset_index()

display(species_bill_length)

for i in species_bill_length.select_dtypes(include=['float', 'int']).columns :
    plt.figure(figsize=(12, 6))
    sns.barplot(data, x='island', y=i, hue='island', palette='Set2')
    plt.show()
    
```
