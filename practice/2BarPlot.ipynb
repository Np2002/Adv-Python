## Bar PLot Using Seaborn 

###  imported all necessary libraries to practicing visualisation

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

### loading dataset from seaborn library dataset 

```python
penguins= sns.load_dataset('penguins')
penguins.head()
```

### checking how many rows and columns are there in the dataset

```python
penguins.shape
```

### now, plotting barplot using seaborn library

```python
sns.barplot(data=penguins, x='island', y='bill_length_mm')
plt.show()
```

### differentiating the graph by male and female

```python
plt.figure(figsize=(14,5))
sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex')
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

### ordering the island based on us how we want to show them in plot

```python
ordered_island = ['Dream', 'Biscoe', 'Torgersen']
plt.figure(figsize=(14,5))
sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex', order=ordered_island)
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

### ordering the gender as female first and then male means we ordering hue values

```python
ordered_island = ['Dream', 'Biscoe', 'Torgersen']
plt.figure(figsize=(14,5))
sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex', order=ordered_island, hue_order=['Female', 'Male'])
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

### changing interval by using "ci" parameter . 

it takes values from 0 to 100.

"ci" deprecated so i used "errorbar=('ci', 0)"

```python
ordered_island = ['Dream', 'Biscoe', 'Torgersen']
plt.figure(figsize=(14,5))
sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex', order=ordered_island, hue_order=['Female', 'Male'],
           errorbar=('ci', 100))
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

### showing interval with some gap using n_boot parameter 

```python
ordered_island = ['Dream', 'Biscoe', 'Torgersen']
plt.figure(figsize=(14,5))
sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex', order=ordered_island, hue_order=['Female', 'Male'],
           errorbar=('ci', 100), n_boot=4)
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

### changing orientation from vertical to horizontal  

in this case "x" should be numeric and "y" should be categorical column.

```python
ordered_island = ['Dream', 'Biscoe', 'Torgersen']
# plt.figure(figsize=(10,6))
sns.barplot(data=penguins, y='island', x='bill_length_mm', hue='sex', order=ordered_island, hue_order=['Female', 'Male'],
           orient='h')
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

### changing bar color to 'green' using parameter "color"

```python
plt.figure(figsize=(14,5))
sns.barplot(data=penguins, x='island', y='bill_length_mm', color='g')
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

### changing color pattern of female and male bars that are differentiates each other using "palette"

```python
plt.figure(figsize=(14,5))
sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex', palette='Set1')
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

### using saturation - for lighting and bluring color palette

```python
plt.figure(figsize=(14,5))
sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex', palette='Set1', saturation=0.3)
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

```python
for i in range(0, 30, 5):
    plt.figure(figsize=(14,5))
    sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex', palette='Set1', saturation=i)
    plt.title('Bar Plot differeciated with male and female')
    plt.show()

```

### changing colors of interval to yellow

```python
plt.figure(figsize=(14,5))
sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex', palette='Set1', errcolor='y')
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

### changing width of interval 

```python
plt.figure(figsize=(14,5))
sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex', palette='Set1', errcolor='y', errwidth=30)
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

```python
plt.figure(figsize=(14,5))
sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex', palette='Set1', err_kws={'color':'g', 'linewidth':20})
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

### adding cap above below the interval line 

```python
plt.figure(figsize=(14,5))
sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex', palette='Set1', err_kws={'color':'g'}, capsize=0.2)
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

### superimposing male and female bars to each other

```python
plt.figure(figsize=(14,5))
sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex', palette='Set1', err_kws={'color':'g'}, dodge=False)
plt.title('Bar Plot differeciated with male and female')
plt.show()
```

### setting background of chart area 

 white, dark, whitegrid, darkgrid, ticks

```python
for i in ['white', 'dark', 'whitegrid', 'darkgrid', 'ticks']:
    sns.set(style=i) 
    plt.figure(figsize=(14,5))
    sns.barplot(data=penguins, x='island', y='bill_length_mm', hue='sex', palette='Set1', err_kws={'color':'g'})
    plt.title(f'{i} - Bar Plot differeciated with male and female ')
    plt.show()
```
