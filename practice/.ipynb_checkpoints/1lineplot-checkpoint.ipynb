```python
#  imported all necessary libraries to practicing visualisation

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# loaded dataset of penguins from seaborn dataset 
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# ploted the line chart using seaborn 
sns.lineplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='sex')
plt.show()
```

```python
# ploted the line chart using matplotlib
plt.plot(penguins[['bill_depth_mm', 'bill_length_mm']])
plt.show()
```

```python
# used line styles based on sex values
sns.lineplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='sex', style='sex')
plt.show()
```

```python
# used palette - Accent
sns.lineplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='sex', style='sex', palette="Accent")
plt.show()
```

```python
# changes the lines colors by using "palette" - "Dark2"
sns.lineplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='sex', style='sex', palette="Dark2")
plt.show()
```

```python
# added markers for different sex values as per # unique sex values - 1 for male - 'o' circle point,  2 for female - '>' (arrow)
plt.figure(figsize=(12,6))
sns.lineplot(data=penguins.head(100), x='bill_length_mm', y='flipper_length_mm', hue='sex', style='sex', palette="Dark2", markers=["o", ">"])
plt.show()
```

```python
# making dashes line into solid line by using parameter "dashes"
plt.figure(figsize=(12,6))
sns.lineplot(data=penguins.head(100), x='bill_length_mm', y='flipper_length_mm', hue='sex', style='sex', palette="Dark2", markers=["o", ">"], dashes=False)
plt.show()
```

```python
# making invisible to legend by "legend" parameter setting it to "False" Value as Boolean
plt.figure(figsize=(12,6))
sns.lineplot(data=penguins.head(100), x='bill_length_mm', y='flipper_length_mm', hue='sex', style='sex', palette="Dark2",
             markers=["o", ">"], dashes=False, legend=False)
plt.show()
```

```python
# added grid to chart by using  "grid" function setting it to "True"
plt.figure(figsize=(14,5))
sns.lineplot(data=penguins.head(100), x='bill_length_mm', y='flipper_length_mm', hue='sex', style='sex', palette="Dark2",
             markers=["o", ">"], dashes=False, legend=False)
plt.title("Line Plot using Seaborn and Matplotlib", fontsize=15)
plt.grid(True)
plt.show()
```
