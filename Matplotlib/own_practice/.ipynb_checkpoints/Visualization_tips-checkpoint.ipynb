### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(10,5))
x=sns.boxplot(data=tips, x='day', y='total_bill', hue='time')
x.set(title="Total Bill per Day by Time", xlabel="Day", ylabel="Total Bill")
# sns.despine()
plt.grid(True)
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
x=sns.histplot(data=tips[tips['time']=='Dinner'], x='tip', bins=10)
x.set(xlabel='Tips', ylabel='Frequency', title='Distribution of tip based time Dinner')
# label = range(11)
# plt.xticks(label)
x=sns.histplot(data=tips[tips['time']=='Lunch'], x='tip', bins=10)
x.set(xlabel='Tips', ylabel='Frequency', title='Distribution of tip based time Lunch')
plt.show()
```

### BONUS: Create your own question and answer it using a graph.

```python
student = ['r', 'b', 'y', 'g']
marks = [20, 90, 48, 54]
```

```python
plt.plot(student, marks, color='r', marker='o')
plt.xlabel('students')
plt.ylabel('Marks')
plt.xticks(student)
plt.yticks(marks)
plt.title("Student's marks ")
plt.grid(True)
plt.show()
```

```python
sns.lineplot(x=student, y=marks)
```

```python
plt.bar(student, marks, width=0.5)
plt.show()
```

```python
sns.barplot(x=student,y=marks)
```
